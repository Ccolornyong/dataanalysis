# 1. 프로젝트 설명
### 프로젝트 명: 주택 가격 예측
### 분석의 목적: 
최근 부동산 업계에서 전세사기 및 허위매물의 피해가 늘고 있다.  
이러한 사기 매물은 시세보다 저렴한 가격으로 매매 사이트에 올라온다는 특징이 있으며 구매자는 이러한 거짓정보에 속아 시간적, 금전적으로 원치않던 주택 계약, 기존보다 높은 가격에 매매, 문제가 있는 주택 계약 등의 피해를 보게 된다. <br>이에 '부동산관고시장감시센터'에서는 예방책으로 "시세보다 가격이 너무 낮다면 의심해 봐야한다"라는 안내를 주고 있지만 구매자가 짧은 시간에 특정 지역의 여러 요소들을 고려해서 적정 시세를 파악하는 것은 매우 어렵다.<br>
따라서, 이런 피해를 방지하고자 여러 요인들을 분석하여 주택가격을 예측하는 모델을 제작하고자 한다. <br>한국보다 다양한 변수들이 있는 미국 데이터를 사용하였으며 이후 안정적인 성능이 도출된다면 한국 기반 데이터를 활용하여 맞춤화 시킬 예정이다. 이를 통해 구매자는 적정 가격 정보를 예측 받아 안정한 부동산 매매에 도움을 받을 수 있을 것이다.<br>
또한, 데이터 분석에서 자주 쓰이는 주택 가격 데이터를 통해 여러 데이터 분석 기술을 익히고 주택 가격에 대해 심층적인 분석을 하고자 한다.
